```python
# Problem 1: Basic Plotting with Matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # assigns the x values
y = np.sin(x) # assigns the y values

plt.figure(figsize=(8, 5))
plt.plot(x, y, linestyle='--', color='blue')  # gives a dashed blue line
plt.title('Trend of sin(X)')  # creates title
plt.xlabel('X')  # X-axis label
plt.ylabel('sin(X)')  # Y-axis label

plt.show()
```

```python
# Problem 2: Exploring Distributions with Seaborn
import seaborn as sns

tips = sns.load_dataset("tips") # load in the tips dataset

# Task 1
plt.figure(figsize=(8, 5)) # creates figure for histogram
sns.histplot(tips['total_bill'], bins=20, kde=False, color='skyblue') # ensures histogram has 20 bins and KDE is false
plt.title('Distribution of Total Bill')  # creates title
plt.xlabel('Total Bill')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.show()

# Task 2
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='orange')
plt.title('Distribution of Total Bill with KDE')  # Title
plt.xlabel('Total Bill')  # X-axis label
plt.ylabel('Density')  # Y-axis label
plt.show()
```

```python
# Problem 3: Categorical Data Analysis

fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # creates subplot

# Box Plot
sns.boxplot(x='time', y='tip', data=tips, ax=axes[0])
axes[0].set_title('Box Plot of Tips by Time of Day')  # Title
axes[0].set_xlabel('Time of Day')  # X-axis label
axes[0].set_ylabel('Tip Amount')  # Y-axis label

# Violin Plot
sns.violinplot(x='time', y='tip', data=tips, ax=axes[1])
axes[1].set_title('Violin Plot of Tips by Time of Day')  # Title
axes[1].set_xlabel('Time of Day')  # X-axis label
axes[1].set_ylabel('Tip Amount')  # Y-axis label

plt.tight_layout()
plt.show()
```
